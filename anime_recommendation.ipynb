```python
import time
import requests
import json
import pandas as pd

# Define constants
API_URL = "https://api.myanimelist.net/v2/anime/ranking"
HEADERS = headers  # Replace with your headers
LIMIT = 500
MAX_RESULTS = 15000

# Function to fetch anime data
def fetch_anime_data():
    all_anime = []
    # Specify the additional fields you want
    fields = "id,title,genres,popularity,synopsis,mean,rank"
    for offset in range(0, MAX_RESULTS, LIMIT):
        # Build the URL with the current offset and fields
        url = f"{API_URL}?ranking_type=all&limit={LIMIT}&offset={offset}&fields={fields}"
        print(f"Fetching: {url}")
        response = requests.get(url, headers=HEADERS)

        if response.status_code == 200:
            data = response.json()
            all_anime.extend(data.get("data", []))  # Append current batch of anime
        else:
            print(f"Error {response.status_code}: {response.text}")
            break  # Stop fetching if there's an error

        # Rate limit handling (optional)
        time.sleep(1)  # Avoid hitting API rate limits

    return all_anime

# Fetch and save data
anime_data = fetch_anime_data()
print(f"Retrieved {len(anime_data)} anime.")
```

```python
import numpy as np

def create_anime_features(anime_df_synopsis, genre_df, ranking_df, n_samples=None):
    # Determine the minimum length of samples across the three inputs
    min_samples = min(len(anime_df_synopsis), len(genre_df), len(ranking_df))

    # If n_samples is specified and is less than min_samples, use it; otherwise, use min_samples
    if n_samples is not None:
        n_samples = min(n_samples, min_samples)
    else:
        n_samples = min_samples

    # Truncate all components to match n_samples
    anime_df_synopsis = anime_df_synopsis[:n_samples]
    genre_df = genre_df.head(n_samples)
    ranking_df = ranking_df.head(n_samples)

    # Ensure genre_df contains only genre columns
    genre_columns = [col for col in genre_df.columns if col != 'id' and col != 'title']

    # Create combined DataFrame with genre and ranking columns
    combined_df = genre_df.copy()
    combined_df['ranking_normalized'] = ranking_df['ranking_normalized'].values

    # Create features column using numpy
    def create_feature_vector(row):
        genre_values = row[genre_columns].values
        ranking_value = [row['ranking_normalized']]
        synopsis_vector = anime_df_synopsis[row.name] if row.name < len(anime_df_synopsis) else np.zeros(len(anime_df_synopsis[0]))

        return np.concatenate([genre_values, ranking_value, synopsis_vector])

    combined_df['features'] = combined_df.apply(create_feature_vector, axis=1)

    return combined_df
```

```python
import time
import torch

# Convert features to PyTorch tensors
anime_features = torch.tensor(list(anime_df['features'].values.tolist()), dtype=torch.float32)
user_features = torch.tensor(list(user_df['features'].values.tolist()), dtype=torch.float32)

# If you have a GPU, move tensors to the GPU for faster processing
if torch.cuda.is_available():
    anime_features = anime_features.cuda()
    user_features = user_features.cuda()

# Initialize an empty DataFrame to store recommendations
recommendations = pd.DataFrame(columns=anime_df.columns)

# Start time tracking
start_time = time.time()

# Loop through each user and calculate cosine similarity with all anime
for user_idx, user_row in user_df.iterrows():
    # Get the feature vector for the current user
    user_feature_vector = user_features[user_idx].unsqueeze(0)  # Add batch dimension

    # Calculate cosine similarities using PyTorch
    similarities = torch.nn.functional.cosine_similarity(user_feature_vector, anime_features)

    # Convert similarities back to a DataFrame-compatible format (CPU and numpy)
    similarities = similarities.cpu().numpy()
    anime_df['similarity'] = similarities

    # Select top 5 recommendations (excluding already-watched animes)
    top_recommendations = anime_df[~anime_df['id'].isin(user_df['id'])].nlargest(5, 'similarity')
    recommendations = pd.concat([recommendations, top_recommendations])

# End time tracking
end_time = time.time()
print(f"Execution time: {end_time - start_time} seconds")

# Drop duplicates and sort recommendations
recommendations = recommendations.drop_duplicates(subset='id').sort_values(by='similarity', ascending=False)
```
